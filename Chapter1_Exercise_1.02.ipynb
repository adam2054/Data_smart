In this exercise you will implement some basic numerical operations with Pandas

```python
import pandas as pd
import numpy as np
import random
```

```python
# Create three series 
np.random.seed(123)
ser1 = pd.Series(np.random.normal(3.0, 1, 100))
ser2 = pd.Series(np.random.normal(5.0, 3, 100))
ser3 = pd.Series(np.random.normal(1.0, 0.5, 100))
```

```python
# Concatenate all the series into a pandas data frame
Df = pd.concat([ser1,ser2,ser3], axis=1)
# Name the columns
Df.columns=['V1','V2','V3']
# Display the head of the column
Df.head()
```

```python
# Calculate the sum of the columns
Df.sum(axis=0)
```

```python
# Find the mean of the columns
Df.mean(axis=0)
```

```python
# Find the mean accross the rows
Df.mean(axis=1)
```

```python
# Apply the divmod function to each of the series
Df.apply(lambda x: divmod(x,3))
```

```python
# Applymap function to generate the modulus of each element in the data frame
Df.applymap(lambda x: divmod(x,3))
```

```python
# Create a list of character 
list1 = [['20']*10,['35']*15,['40']*10,['10']*25,['15']*40]
# Convert them into a single list using list comprehension
charlist = [x for sublist in list1 for x in sublist]
# Display the output
len(charlist)
```

```python
# Randomly shuffle the character list
random.seed(123)
random.shuffle(charlist)
# Convert the list to a series
ser4 = pd.Series(charlist)
ser4
```

```python
# Converting the series to numeric
ser4 = pd.to_numeric(ser4)
ser4
```

```python
# Concatinating the new series with the old data frame
Df = pd.concat([Df,ser4],axis=1)
# Renaming the data frame
Df.rename(columns={0:'V4'}, inplace=True)
# Displaying the data frame
Df
```

```python
# Converting to a numpy array
numpArray = Df.to_numpy()
# Displaying the numpy array
numpArray
```
